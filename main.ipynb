```
from anthropic import Anthropic
client = Anthropic(api_key="")
MODEL_NAME = "claude-3-haiku-20240307"
```

```
def get_completion(client, prompt):
    return client.messages.create(
        model=MODEL_NAME,
        max_tokens=2048,
        messages=[{
            "role": 'user', "content":  prompt
        }]
    ).content[0].text
```

```
completion = get_completion(client,
    f"""

Please do the following:
1. Which is the most strongest heroes ?
"""
)
print(completion)
```

```
completion2 = get_completion(client,
    f"""Here is an hero details file: <file>{content}</file>

Do this following:
1. how many hero in that file , list the name?
2. Which is the most strongest heroes ?
3. If i want to play karina. which hero in the list that i will need to be careful of when fighting ?
4. Let say my enemy like to play an a high mobility hero , which heroes that i can use to counter that ?
5. Let say my enemy like to play an a tanky hero, which heroes that i can use to counter that ?
"""
)
print(completion2)
```
